```python
import cv2
import numpy as np
import tensorflow as tf
import tensorflow_hub as hub
import pandas as pd
import glob
import pandas as pd
```

```python
module = hub.Module("https://tfhub.dev/google/imagenet/inception_v3/feature_vector/1")
height, width = hub.get_expected_image_size(module)
```

```python
data_temp=np.zeros(shape=(100,299,299,3))
```

```python
for folder_no in range(1,101):
    images = []
    count=0
    folder_feat="/home/sachin/ML_training/Dog_breed/desti/"+str(folder_no)+"/*.jpg"
    for img in glob.glob(folder_feat):
        n= cv2.imread(img)
        images.append(n)
        count=count+1
        if count==100:
            break
    images=np.array(images)
    for i_s in range(images.shape[0]):
        data_temp[i_s]=cv2.resize(images[i_s],(height,width))/255
        
    #Running our session
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        features=module(data_temp)
        dat1=features.eval()
    df = pd.DataFrame(dat1)
    feat_file_name="feat_file_"+str(folder_no)+".csv"
    df.to_csv(feat_file_name)
```
